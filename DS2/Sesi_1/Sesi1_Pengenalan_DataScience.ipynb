## 1. Apa itu Data Science?

**Data Science** adalah bidang interdisiplin yang menggabungkan statistika, ilmu komputer, dan pengetahuan domain untuk mengekstrak wawasan (*insight*) dari data, baik data terstruktur maupun tidak terstruktur.

Tiga pilar utama Data Science:
1. **Matematika & Statistika** – dasar analisis kuantitatif
2. **Ilmu Komputer & Pemrograman** – pengolahan data dalam skala besar
3. **Pengetahuan Domain (Business Acumen)** – memahami konteks masalah

## 2. Siklus Hidup Data Science (Data Science Life Cycle)

| Tahap | Deskripsi |
|---|---|
| 1. Business Understanding | Memahami masalah dan tujuan bisnis |
| 2. Data Collection | Mengumpulkan data dari berbagai sumber |
| 3. Data Cleaning/Preprocessing | Membersihkan data dari noise dan error |
| 4. Exploratory Data Analysis (EDA) | Eksplorasi pola dan hubungan dalam data |
| 5. Modeling | Membangun model statistik/machine learning |
| 6. Evaluation | Mengukur performa model |
| 7. Deployment | Menerapkan model ke sistem produksi |

## 5. Jenis-Jenis Data

- **Data Terstruktur (Structured)** → tabel, spreadsheet, database (mudah diolah)
- **Data Semi-Terstruktur (Semi-structured)** → JSON, XML
- **Data Tidak Terstruktur (Unstructured)** → gambar, teks bebas, audio, video

```python
# Contoh data terstruktur sederhana
data_terstruktur = pd.DataFrame({
    'nama'   : ['Andi', 'Budi', 'Citra'],
    'usia'   : [21, 22, 20],
    'jurusan': ['Informatika', 'Statistika', 'Matematika']
})
print('Contoh Data Terstruktur:')
print(data_terstruktur)

# Contoh data semi-terstruktur (JSON)
import json
data_json = {
    'nama': 'Andi',
    'kontak': {'email': 'andi@mail.com', 'telepon': '08123456789'},
    'hobi': ['membaca', 'coding']
}
print('\nContoh Data Semi-Terstruktur (JSON):')
print(json.dumps(data_json, indent=2))
```

## 7. Contoh Studi Kasus Sederhana

Misalkan kita memiliki data penjualan harian dan ingin tahu rata-rata penjualan.

```python
penjualan_harian = [120, 135, 98, 150, 110, 175, 200]
hari = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']

df_penjualan = pd.DataFrame({'Hari': hari, 'Unit Terjual': penjualan_harian})
print(df_penjualan)

print(f'\nRata-rata penjualan harian : {sum(penjualan_harian)/len(penjualan_harian):.2f} unit')
print(f'Hari penjualan tertinggi    : {hari[penjualan_harian.index(max(penjualan_harian))]} ({max(penjualan_harian)} unit)')
print(f'Hari penjualan terendah     : {hari[penjualan_harian.index(min(penjualan_harian))]} ({min(penjualan_harian)} unit)')
```
